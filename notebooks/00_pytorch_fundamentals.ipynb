## Random tensors 

```python
# Create a random tensor in Pytorch

random_tensor = torch.rand(3,4)
random_tensor
```

```python
random_tensor.ndim
```

```python
random_tensor.shape
```

```python
# Random tensor to similar shape as image tensor

random_image_tensor = torch.rand(size=(224, 224, 3))
random_image_tensor.shape, random_image_tensor.ndim
```

## Zeros and ones

```python
# Tensor of all zeros

zeros = torch.zeros(3,4)
zeros
```

```python
zeros*random_tensor
```

```python
# Tensor of all ones

ones = torch.ones(3,4)
ones
```

## Creating a range of tensors and tensors-like

```python
# Creating a range

one_to_ten = torch.arange(start=1, end=11, step=1)
one_to_ten
```

```python
# Tensors like

ten_zeros = torch.zeros_like(input=one_to_ten)
ten_zeros
```

## Tensor data type

```python
# float32 tensor

float32_tensor = torch.tensor([3., 6., 9.], 
                              dtype=None,
                              device=None,
                              requires_grad=False)
float32_tensor
```
